```python
import random
import matplotlib.pyplot as plt
import numpy as np
from sklearn.isotonic import IsotonicRegression
```

```python
def fix(x, y, y_ir, length):
    i = 0
    while i < len(y_ir):
        if abs(y_ir[i] - y[i]) > length:
            y_ir = np.delete(y_ir, i)
            y = np.delete(y, i)
            x = np.delete(x, i)
            i -= 1
        i += 1
    return x, y, y_ir
```

```python
x = sorted(random.choices(range(1, 51), k=100))
y = [i*np.random.uniform(1,2)+np.random.uniform(-10,10) for i in x]
x = np.array(x)
y = np.array(y)
plt.scatter(x,y)
plt.show()
```

```python
ir = IsotonicRegression()
ir.fit(x,y)
y_ir = ir.fit_transform(x, y)
plt.scatter(x,y)
plt.plot(x,y_ir,c = 'red')
plt.show()
```

```python
for i in range(10):
    x, y, y_ir = fix(x, y, y_ir, 10 - i)
    ir = IsotonicRegression()
    ir.fit(x,y)
    y_ir = ir.fit_transform(x, y)
    plt.scatter(x,y)
    plt.plot(x,y_ir,c = 'red')
    plt.show()
```
