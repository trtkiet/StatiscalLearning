```python
#Import thư viện
import numpy as np
from matplotlib import pyplot as plt
```

```python
#Sinh data 
n = 100
x = np.linspace(0.1, 10, n)
y_smooth = x**3 + x**2 - 2*x + 1
s = np.random.normal(0, 50, n)
y = y_smooth + s
```

```python
ax = plt.subplot()
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.plot(x, y, 'o')
```

```python
from GPSymbolicRegression import *

myModel = GeneticSymbolicRegressor(generations=30, population_size=50)
myModel.fit(x, y)
y_predict1 = myModel.predict(x)
```

```python
from pysr import PySRRegressor
pysrModel = PySRRegressor(population_size=50, niterations=30)
x = x.reshape(100, 1)
pysrModel.fit(x, y)
y_predict2 = pysrModel.predict(x)
```

```python
plt.xlabel('X')
plt.ylabel('Y')
plt.plot(x, y, color='b', label='target', marker='o', linestyle='none')
plt.plot(x, y_predict1, color='r', label='simpleGPSR predict')
plt.plot(x, y_predict2, color='g', label='PySR predict')
plt.legend(['Y', 'SimpleGPSR', 'PySR'])
plt.show()
```
